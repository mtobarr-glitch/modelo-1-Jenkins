```python
import os
import cv2
import numpy as np
import pandas as pd
from pathlib import Path

# Rutas
ruta_original = Path(r"C:\Users\PC\Desktop\indice de jenkins\fotos originales\4_semana")  # Ruta de las imágenes originales
ruta_recortada = Path(r"C:\Users\PC\Desktop\indice de jenkins\base de datos\4_semana")  # Ruta de las imágenes recortadas

# Parámetros de la imagen original
diametro_campo_mm = 1.5  # en mm (diámetro del campo visual en el microscopio)

# Función para detectar el diámetro de un círculo en la imagen
def detectar_circulo(imagen):
    if imagen is None:
        raise ValueError("No se pudo cargar la imagen. Verifique la ruta y el nombre del archivo.")
    
    # Convertir la imagen a escala de grises
    gris = imagen
    
    # Aplicar el desenfoque para reducir el ruido
    gris = cv2.GaussianBlur(gris, (15, 15), 0)
    
    # Detectar círculos usando la transformada de Hough
    circulos = cv2.HoughCircles(gris, cv2.HOUGH_GRADIENT, dp=1.2, minDist=20, param1=50, param2=30, minRadius=50, maxRadius=200)

    if circulos is not None:
        circulos = np.round(circulos[0, :]).astype("int")
        # Retornar el diámetro del primer círculo detectado
        return circulos[0][2] * 2  # El diámetro es el doble del radio
    else:
        return None

# Listar las imágenes en la carpeta de las imágenes originales y recortadas
exts = {".jpg", ".jpeg", ".png", ".tif", ".tiff", ".bmp"}
archivos_originales = [f for f in ruta_original.iterdir() if f.suffix.lower() in exts]
archivos_recortados = [f for f in ruta_recortada.iterdir() if f.suffix.lower() in exts]

# Preparar la lista para almacenar los resultados
resultados = []

# Procesar las imágenes originales para obtener el diámetro del campo visual en píxeles
for archivo_original in archivos_originales:
    imagen_original = cv2.imread(str(archivo_original), cv2.IMREAD_GRAYSCALE)  # Leer imagen en escala de grises

    # Verificar que la imagen se haya cargado correctamente
    if imagen_original is None:
        print(f"Error al cargar la imagen original {archivo_original.name}.")
        continue

    diametro_campo_original_px = detectar_circulo(imagen_original)

    # Verificar que se detectó el círculo en la imagen original
    if diametro_campo_original_px is None:
        print(f"No se pudo detectar el campo visual en la imagen original {archivo_original.name}")
        continue

    # Procesar cada imagen recortada correspondiente a la imagen original
    for archivo_recortado in archivos_recortados:
        if archivo_recortado.name == archivo_original.name:  # Asegurarnos de procesar la imagen recortada correspondiente
            # Leer la imagen recortada
            img_recortada = cv2.imread(str(archivo_recortado), cv2.IMREAD_GRAYSCALE)

            # Verificar que la imagen recortada se haya cargado correctamente
            if img_recortada is None:
                print(f"Error al cargar la imagen recortada {archivo_recortado.name}.")
                continue

            # Detectar el diámetro del campo visual en píxeles en la imagen recortada
            diametro_campo_recortado_px = detectar_circulo(img_recortada)

            # Verificar que se detectó el círculo en la imagen recortada
            if diametro_campo_recortado_px is None:
                print(f"No se pudo detectar el campo visual en la imagen recortada {archivo_recortado.name}")
                continue

            # Calcular la escala en mm/px
            escala_mm_px = diametro_campo_mm / diametro_campo_recortado_px  # Usamos el diámetro del campo visual

            # Agregar los resultados
            resultados.append({
                'Imagen Original': archivo_original.name,
                'Imagen Recortada': archivo_recortado.name,
                'Ancho Recortado (px)': img_recortada.shape[1],
                'Alto Recortado (px)': img_recortada.shape[0],
                'Escala (mm/px)': escala_mm_px
            })

# Crear un DataFrame con los resultados
df_resultados = pd.DataFrame(resultados)

# Cambiar la ruta de salida a la nueva ubicación
out_xlsx = Path(r"C:\Users\PC\Desktop\indice de jenkins\filtro erosion\4_semana\escala_mm_px.xlsx")
df_resultados.to_excel(out_xlsx, index=False)

print(f"✅ El archivo Excel con las escalas ha sido guardado en: {out_xlsx}")
```
