```python
from pathlib import Path
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2

# Carpeta con tus imágenes
ruta_base = Path(r"C:\Users\PC\Desktop\materia 2025\segundo semestre\tesis\indice de jenkins\base de datos\1_semana")
# Carpeta para guardar resultados
ruta_salida = ruta_base / "procesadas_erosion"
ruta_salida.mkdir(exist_ok=True)

# Kernel para erosión 2x2
kernel = np.ones((2, 2), np.uint8)

# Obtener automáticamente todas las imágenes .jpg y .png de la carpeta
archivos = [f for f in ruta_base.iterdir() if f.suffix.lower() in [".jpg", ".png", ".jpeg"]]

for archivo in archivos:
    # Leer imagen con matplotlib para evitar problemas de OpenCV
    img_color = mpimg.imread(archivo)
    
    # Convertir a escala de grises si tiene color
    if len(img_color.shape) == 3:
        img_gray = cv2.cvtColor((img_color * 255).astype(np.uint8), cv2.COLOR_RGB2GRAY)
    else:
        img_gray = (img_color * 255).astype(np.uint8)

    # Aplicar erosión dos veces
    eroded = cv2.erode(img_gray, kernel, iterations=2)

    # Aplicar filtro mediano (tamaño de kernel 3x3)
    median = cv2.medianBlur(eroded, ksize=3)

    # Guardar resultado
    nombre_salida = ruta_salida / f"{archivo.stem}_eroded_median{archivo.suffix}"
    cv2.imwrite(str(nombre_salida), median)
    
    print(f"✅ Procesada: {archivo.name} → {nombre_salida.name}")

    # Mostrar antes y después de la primera imagen
    if archivo == archivos[0]:
        plt.figure(figsize=(12,5))
        plt.subplot(1,3,1)
        plt.imshow(img_gray, cmap='gray')
        plt.title("Original")
        plt.axis("off")

        plt.subplot(1,3,2)
        plt.imshow(eroded, cmap='gray')
        plt.title("Después de erosión 2x2")
        plt.axis("off")

        plt.subplot(1,3,3)
        plt.imshow(median, cmap='gray')
        plt.title("Después de filtro mediano")
        plt.axis("off")

        plt.show()
```
